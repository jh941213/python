## 1. 머쓱이보다 키 큰 사람

image.png

```python
import time 

start = time.time()


def solution(array, height):
# 답안 작성 부분 ===============
    answer = 0
    for i in array:
        if i > height:
            answer += 1 

# ==============================

    return answer

print(solution([149, 180, 192, 170], 167))
print(solution([180, 120, 140], 190))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 약수 구하기

image.png

```python
import time 

start = time.time()


def solution(num):
# 답안 작성 부분 ===============
    answer=[]
    for i in range(1,num+1):
        if num % i == 0:
            answer.append(i)




# ==============================

    return answer

print(solution(24))
print(solution(29))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 4. 7의 개수

image.png

```python
import time 

start = time.time()


def solution(num):
# 답안 작성 부분 ===============
    answer =[]
    for i in num:
        answer+=str(i)
# ==============================

    return answer.count('7')

print(solution([7, 77, 17]))
print(solution([10, 29]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
